```
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv(url, error_bad_lines=False)

#Plots in matplotlib reside within a figure

fig=plt.figure()


ax = fig.add_subplot(1,1,1)

#Variable
ax.hist(df['Healthy life expectancy'],bins = 5)
#Labels and Tit
plt.title('Jumlah Kesehatan dan Korupsi')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Perceptions of corruption')
plt.show()
```

**Analisis Histogram**

Pada data yang telah saya analisis menunjukkan diagram histogram yang tidak normal, pada data kategori harapan hidup sehat data menunjukkan tertinggi 0.8 Jadi jumlah harapan hidup sehat di suatu negara tidak merata yang dihasilkan dengan total data keseluruhan

```
fig=plt.figure()

ax = fig.add_subplot(1,1,1)

ax.scatter(df['Healthy life expectancy'],df['Perceptions of corruption'])

plt.title('Data Harapan Hidup Sehat dan Persepsi korupsi')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Perceptions of corruption')
plt.show()
```

**Analisis Scatter Plot**

Pada data hasil dari scratter plot yang telah saya analisis menunjukkan korelasi data antara Healthy life expectancy dengan Perceptions of corruption yang saya inputkan memiliki tidak memiliki hubungan secara linear artinya semakin tinggi Healthy life expectancy tidak selalu mempengaruhi Perceptions of corruption

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(url, error_bad_lines=False)

df = df.pivot_table(index='Healthy life expectancy', columns='Perceptions of corruption', values='Social support')

print (df)

sns.heatmap(df, annot=True, cmap='coolwarm')

plt.show()
```
